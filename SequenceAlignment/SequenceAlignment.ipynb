```python
# Parses fasta file
def read_fasta_file(inp_fasta_file):
    name, seq = '', []
    for line in inp_fasta_file:
        line = line.rstrip()
        if line.startswith(">"):
            if name:
                yield(name, ''.join(seq))
            name, seq = line, []
        else: 
            seq.append(line)
    if name: 
        yield(name, ''.join(seq))
```

```python
def needleman_wunsch(d_same, d_replace, d_open, d_extension, seq1, seq2):
    M = [[float('-inf')] * (len(seq2) + 1) for i in range(len(seq1) + 1)] 
    Del = [[float('-inf')] * (len(seq2) + 1) for i in range(len(seq1) + 1)] 
    Ins = [[float('-inf')] * (len(seq2) + 1) for i in range(len(seq1) + 1)] 
    
    M[0][0] = 0
    Del[0][0] = 0
    Ins[0][0] = 0
    for i in range(1, len(seq1) + 1):
        Del[i][0] = d_open + (i - 1) * d_extension
    for j in range(1, len(seq2) + 1):
        Ins[0][j] = d_open + (j - 1) * d_extension
        
    for i in range(1, len(seq1) + 1):
        for j in range(1, len(seq2) + 1):
            if seq1[i - 1] == seq2[j - 1]:
                d = d_same
            else: 
                d = d_replace
            Del[i][j] = max(Del[i - 1][j] + d_extension, M[i - 1][j] + d_open)
            Ins[i][j] = max(Ins[i][j - 1] + d_extension, M[i][j - 1] + d_open)
            M[i][j] = max(Del[i][j], Ins[i][j], M[i - 1][j - 1] + d)
                
    print(round(M[-1][-1], 3))
          
    res1, res2 = [], []
    i, j = len(seq1), len(seq2)
    score = M[-1][-1]
    state = 'M'
    
    while score != 0:
        if state == 'M':
            if seq1[i - 1] == seq2[j - 1]:
                d = d_same
            else: 
                d = d_replace
            
            if M[i - 1][j - 1] + d == score:
                state = 'M'
                p = (-1, -1)
                score = M[i - 1][j - 1]
            elif Del[i][j] == score:
                state = 'D'
            elif Ins[i][j] == score:
                state = 'I'
                
        if state == 'D':
            p = (-1, 0)
            if Del[i - 1][j] + d_extension == score:
                state = 'D'
                score = Del[i - 1][j]
            elif M[i - 1][j] + d_open == score:
                state = 'M'
                score = M[i - 1][j]
        elif state == 'I':
            p = (0, -1)
            if Ins[i][j - 1] + d_extension == score:
                state = 'I'
                score = Ins[i][j - 1]
            elif M[i][j - 1] + d_open == score:
                state = 'M'
                score = M[i][j - 1]
        
        if p == (-1, 0):
            res1.append(seq1[i - 1])
            res2.append('-')
            i -= 1
        elif p == (0, -1):
            res1.append('-')
            res2.append(seq2[j - 1])
            j -= 1
        elif p == (-1, -1):
            if seq1[i - 1] != seq2[j - 1]:
                res1.append(seq1[i - 1].lower())
                res2.append(seq2[j - 1].lower())
            else:
                res1.append(seq1[i - 1])
                res2.append(seq2[j - 1])
            i, j = i - 1, j -1 
    res1.reverse()
    res2.reverse()
    print(''.join(res1))
    print(''.join(res2))
    return ''.join(res1), ''.join(res2)
            
import sys
from Bio import SeqIO

def main():
    sys.stdin = open("test_in.txt", "r")
    sys.stdout = open("test_out.txt", "w")
    input = sys.stdin.readline

    d_same, d_replace, d_open, d_extension = list(map(float, input().split()))
            
    sequences = [record.seq for record in SeqIO.parse(sys.stdin, "fasta")]
    seq1, seq2 = sequences[0], sequences[1]
    
    needleman_wunsch(d_same, d_replace, d_open, d_extension, seq1, seq2)
main()
```
